```python
# preparing a column, presenting the max number of orders for every customer
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merged.head(100)
```

```python
#creating a loyality flag
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merged
```

# QN5)

```python
#calculating the aggregated mean,min and max prices by loyalty_flag
ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean',"min","max"]})
```

It's clear to see, that the probable mistake of the max price mentionned in the last exercice is influencing the mean of the prices.

```python
#calculating the mean, min and max of the prices of the different loyality flag customers in the new subset
ords_prods_merged_new.groupby('loyalty_flag').agg({'prices': ['mean',"min","max"]})
```

```python
#creating the column mean_price for mean by different loyality flag customers
ords_prods_merged_new['mean_prices'] = ords_prods_merged_new.groupby(['loyalty_flag'])['prices'].transform(np.mean)
```

```python
ords_prods_merged_new["mean_prices"].value_counts()
```

```python
#creating a column to characterize customers depending on their spendings

ords_prods_merged_new.loc[ords_prods_merged_new['mean_prices'] < 10, 'spender_flag'] = 'low_spender'
ords_prods_merged_new.loc[ords_prods_merged_new['mean_prices'] >= 10, 'spender_flag'] = 'High_spender'
```

```python
ords_prods_merged_new.head(3)
```

# QN7)

```python
# droping all the missing values including the column days since prior order for this task
ords_prods_merged_new_clean= ords_prods_merged_new.dropna()
```

```python
# creating a column for median of days since prior order by users id
ords_prods_merged_new_clean['median_days_since_prior_order'] = ords_prods_merged_new_clean.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# creating a column that characterize customers onhand of their ordering frequency ( time since last order)
ords_prods_merged_new_clean.loc[ords_prods_merged_new_clean['median_days_since_prior_order'] > 20, 'Customer_frequency'] = 'Non-frequent customer'
ords_prods_merged_new_clean.loc[(ords_prods_merged_new_clean['median_days_since_prior_order'] >10) & (ords_prods_merged_new_clean['median_days_since_prior_order'] <= 20), 'Customer_frequency'] = 'Regular customer'
ords_prods_merged_new_clean.loc[ords_prods_merged_new_clean['median_days_since_prior_order'] <= 10, 'Customer_frequency'] = 'Frequent customer'
```
